### Cleaning Data

####Untuk memudahkan analisis berbasis waktu maka kolom year, month, day dan hour digabung menjadi satu kolom datetime

```python
def create_datetime_column(df):
    """Combines year, month, day, and hour columns into a single datetime column.

    Args:
        df: pandas DataFrame containing 'year', 'month', 'day', and 'hour' columns.

    Returns:
        pandas DataFrame with an added 'datetime' column.
    """
    try:
        df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
        return df
    except KeyError as e:
        print(f"Error: Column {e} not found in DataFrame.")
        return df

# Apply the function to each DataFrame
Aotizhongxin_df = create_datetime_column(Aotizhongxin_df)
Changping_df = create_datetime_column(Changping_df)
Dingling_df = create_datetime_column(Dingling_df)
Dongsi_df = create_datetime_column(Dongsi_df)
Guanyuan_df = create_datetime_column(Guanyuan_df)
Gucheng_df = create_datetime_column(Gucheng_df)
Huairou_df = create_datetime_column(Huairou_df)
Nongzhanguan_df = create_datetime_column(Nongzhanguan_df)
Shunyi_df = create_datetime_column(Shunyi_df)
Tiantan_df = create_datetime_column(Tiantan_df)
Wanliu_df = create_datetime_column(Wanliu_df)
Wanshouxigong_df = create_datetime_column(Wanshouxigong_df)
```

####Menghapus kolom yang dianggap tidak perlu seperti kolom No, year, month, day dan hour

```python
def drop_columns(df):
    cols_to_drop = ['No', 'year', 'month', 'day', 'hour']
    df = df.drop(columns=cols_to_drop, errors='ignore')
    return df

# Apply the function to each DataFrame
Aotizhongxin_df = drop_columns(Aotizhongxin_df)
Changping_df = drop_columns(Changping_df)
Dingling_df = drop_columns(Dingling_df)
Dongsi_df = drop_columns(Dongsi_df)
Guanyuan_df = drop_columns(Guanyuan_df)
Gucheng_df = drop_columns(Gucheng_df)
Huairou_df = drop_columns(Huairou_df)
Nongzhanguan_df = drop_columns(Nongzhanguan_df)
Shunyi_df = drop_columns(Shunyi_df)
Tiantan_df = drop_columns(Tiantan_df)
Wanliu_df = drop_columns(Wanliu_df)
Wanshouxigong_df = drop_columns(Wanshouxigong_df)
```

#####Membandingkan nilai polutan dengan nilai standar WHO

#####Melihat tingkat polutan yang melebihi standar WHO berdasarkan siang dan malam

```python
# Define day as 6 AM to 6 PM (inclusive), otherwise it's night.
df['time_of_day'] = df['datetime'].dt.hour.apply(lambda x: 'day' if 6 <= x <= 18 else 'night')

# Group the data by time of day and pollutant, and calculate the mean.
pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
for pollutant in pollutants:
    day_night_means = df.groupby('time_of_day')[pollutant].mean()
    print(f"Average {pollutant} levels by time of day:")
    print(day_night_means)
    print("-" * 20)

# Further analysis: Comparing exceedances of WHO standards
who_standards = {
    'PM2.5': 5,  # g/m annual mean
    'PM10': 20,  # g/m annual mean
    'SO2': 20,  # g/m annual mean
    'NO2': 25,  # g/m annual mean
    'CO': 4000,  # g/m 8-hour mean
    'O3': 100,  # g/m 8-hour mean
}

for pollutant in pollutants:
    # Calculate exceedance for each time of day
    day_exceedance = df[(df['time_of_day'] == 'day') & (df[pollutant] > who_standards[pollutant])][pollutant].count()
    night_exceedance = df[(df['time_of_day'] == 'night') & (df[pollutant] > who_standards[pollutant])][pollutant].count()
    print(f"Exceedances of {pollutant} WHO standards:")
    print(f"Day: {day_exceedance}")
    print(f"Night: {night_exceedance}")
    print("-" * 20)

```

#####Melihat tingkat signifikansi perbedaan tingkat polutan antara siang dan malam hari

```python
pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']

for pollutant in pollutants:
    # Extract pollutant values for day and night
    day_values = df[df['time_of_day'] == 'day'][pollutant]
    night_values = df[df['time_of_day'] == 'night'][pollutant]

    # Perform the t-test
    t_statistic, p_value = stats.ttest_ind(day_values, night_values)

    print(f"T-test for {pollutant}:")
    print(f"  T-statistic: {t_statistic:.2f}")
    print(f"  P-value: {p_value:.3f}")

    alpha = 0.05  # significance level
    if p_value < alpha:
        print(f"  There is a statistically significant difference in {pollutant} levels between day and night.")
    else:
        print(f"  There is no statistically significant difference in {pollutant} levels between day and night.")
    print("-" * 20)
```
